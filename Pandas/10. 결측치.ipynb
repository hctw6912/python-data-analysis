## 데이터 제외하기 dropna

```python
import pandas as pd

df = pd.read_excel('score.xlsx', index_col = '지원번호')
df
```

```python
df = df.dropna() # Nan 을 포함하는 데이터 삭제
```

```python
df
```

```python
import pandas as pd

df = pd.read_excel('score.xlsx', index_col = '지원번호')
df
```

- axis : index or columns
- how : any or all

```python
df.dropna(axis = 'index', how = 'any')
```

```python
df.dropna(axis = 'columns') # NAn이 하나라도 있는 columns 삭제
```

```python
df['학교'] = np.nan
df
```

```python
df.dropna(axis = 'columns', how = 'all') # 데이터 전체가 Nan 만 경우에만 column 삭제
```
